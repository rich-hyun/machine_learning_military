```python
from google.colab import files
uploaded=files.upload()

import pandas as pd
workbook=pd.read_excel('project_proof2.xlsx')
base=workbook.to_numpy()

import matplotlib.pyplot as plt
plt.scatter(base[:52,[1]],base[:52,[3]],color='c')
plt.scatter(base[52:,[1]],base[52:,[3]],color='m')
plt.scatter(500, 23, marker='^',color='g')
plt.xlabel('remaining days')
plt.ylabel('age')
plt.show()

"""k-  """
age=[]
for k in range(66):
  age.append(base[k][1])
rdays=[]
for k in range(66):
  rdays.append(base[k][3])

base_data=[[l,w] for l,w in zip(age,rdays)]
base_target=[1]*52+[0]*14

from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()

import numpy as np
input_arr=np.array(base_data)
target_arr=np.array(base_target)

np.random.seed(42)
index=np.arange(66)
np.random.shuffle(index)

train_input=input_arr[index[:35]]
train_target=target_arr[index[:35]]

test_input=input_arr[index[35:]]
test_target=target_arr[index[35:]]

import matplotlib.pyplot as plt
plt.scatter(train_input[:,0],train_input[:,1], color='c')
plt.scatter(test_input[:,0],test_input[:,1],color='m')
plt.xlabel('remaining days')
plt.ylabel('age')
plt.show()

kn=kn.fit(train_input,train_target)
kn.score(test_input,test_target)
```
